# 차원수 늘리기
- 이미지 분석(CNN) 레이어를 사용하기 위해서는 필수적으로 알아야 한다.

```
# Tensorflow 기준으로 CNN에 사용되어지는 배열의 형상은 항상 다음과 같다.
arr2 = arr[np.newaxis, :, :, np.newaxis]
arr2.shape
# 출력값 : (1, 3, 4, 1) : (이미지 갯수, 가로 갯수, 세로 갯수, 채널의 수)
```

```
arr2.ndim
```

```
arr2
```

```
arr3 = np.squeeze(arr2)  # 1로 되어있는 차원의 데이터를 삭제해줌
                         # reshape을 이용해도 똑같은 일 할 수 있음
arr3.shape
```

```
arr4 = arr[np.newaxis, ..., np.newaxis]  #  원래 있던 데이터를 가져오겠다
# ... : 원래 데이터 처음부터 끝까지 가져오겠다는 이유
# arr이 가지고 있던 차원에
# 앞뒤로 새로운 차원 추가하는 것

# ...이 아닌 [np.newaxis, :, :, np.newaxis]가 좀 더 명확하게 보이기는 함
arr4.shape
```

## 형상 변환
- reshape()

```
x = np.arange(15).reshape(3, 5)
x.shape
```

```
# ravel : 평탄화 시킨 참조 배열을 만들어 낸다. (View 또는 Reference)
# flatten : 평탄화 시킨 복사 배열을 만들어 낸다. (복사본)

temp = np.ravel(x)  # x : 3*5 행렬이었음
print(temp)
```

```
temp[0] = 100
print(x)
# ravel시킨 객체를 변경시키면 원본 객체가 변경됨
# 참조.
```

```
y = np.arange(15).reshape(3, 5)
```

```
temp2 = y.flatten()  # y를 1차원으로 쭉 펴낸 배열을 새롭게 만들어 낸다.
# 새롭게!!
print(temp2)
print(y)
```

```
temp2[0] = 100
print(temp2)
print(y)
```

# 브로드 캐스팅
- 차원수가 다른 배열끼리의 연산

```
x = np.arange(15).reshape(3, 5)
y = np.random.randn(15).reshape(3, 5)

print(x)
print(y)
```

```
x * y  # 원소의 곱
```

```
x * 3  # 배열 * 스칼라값
# 스칼라값이 브로드 캐스팅된다.
# 모든 원소에 3이 곱해지는 것이 아닌, 스칼라값이 3*4의 형태로 바뀌는 것
```

```
a = np.arange(12).reshape(4, 3)  # (4, 3)
b = np.arange(100, 103)  # (3,)  # 1차원 배열
c = np.arange(1000, 1004)  # (4,)  # 1차원 배열
d = b.reshape(1, 3)  # (1, 3)

print(a.shape)
print(b.shape)
print(c.shape)
print(d.shape)
```

```
a + b
```

```
a + c
# a의 가로는 3개
# c의 가로는 4개라서
# 브로드캐스팅 불가
```

```
a + d
```
