# 신경망

```
X = np.array([1.0, 0.5])  # 입력값 2개
W1 = np.array([
               [0.1, 0.3, 0.5],  # 2개의 입력을 받는 3개의 뉴런 - 1층의 구조
               [0.2, 0.4, 0.6]    
])

B1 = np.array([0.1, 0.2, 0.3])  # 편향
```

```
X.shape, W1.shape, B1.shape
```

```
# 계산 가능
```

```
# 활성화 함수에 들어갈 결과
z1 = np.dot(X, W1) + B1
z1
```

```
# 신경망이 여러개 있는 것 : 딥러닝
```

```
def sigmoid(x):  # 활성화 함수 sigmoid
  return 1 / (1 + np.exp(-x))
```

```
# 활성화 함수까지 거친 3개의 결과
A1 = sigmoid(z1)
A1  # 브로드 캐스팅 일어남
```

```
# 1층의 출력이 2층의 입력
# 2층의 구조
W2 = np.array([[0.1, 0.4],
               [0.2, 0.5],
               [0.3, 0.6]])  # 3개의 입력을 받는 2개의 뉴런
B2 = np.array([0.1, 0.2])

z2 = np.dot(A1, W2) + B2
A2 = sigmoid(z2)
A2
```

```
# 항등함수 : 신경망 회귀에서 많이 사용
def identity_function(x):
    return x

# 3층의 구조
W3 = np.array([[0.1, 0.3],
               [0.2, 0.4]])
B3 = np.array([0.1, 0.2])

z3 = np.dot(A2, W3) + B3

# 출력층에 항등함수, softmax 많이 사용
y = identity_function(z3)
y
```
